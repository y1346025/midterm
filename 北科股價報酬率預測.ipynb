```python
train_price = list(hist.tail(4)['Close'])
train_return = []
for i in range(1, len(train_price)):
  print((train_price[i]-train_price[i-1])/train_price[i-1])
  train_return.append( (train_price[i]-train_price[i-1])/train_price[i-1] )
```

```python
train_return
```

```python
for i in range(0, 5):
  temp = 0
  for j in train_return:
    temp = temp + j
  temp = temp/3
  train_return.append(temp)
  print('Add', temp, 'current length', len(train_return))
```
