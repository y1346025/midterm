```python
import pandas as pd
import os
import re


# 5個正確值
correct_values = [-0.00578, -0.0022, -0.00875, -0.00873, 0.1196]

# 每個學號的準確度
accuracy_scores = {}
```

```python
# 遍歷資料夾中的所有子資料夾
for folder in os.listdir('midterm_assignment/'):
    folder_path = os.path.join('midterm_assignment', folder)

    # 確認是否為資料夾
    if os.path.isdir(folder_path):
        # 收集該資料夾內所有xlsx檔案
        xlsx_files = [f for f in os.listdir(folder_path) if f.endswith('.xlsx')]

        # 檢查是否恰好有一個xlsx檔案
        if len(xlsx_files) == 1:
            file = xlsx_files[0]
            # 擷取學號，檔名：學號_姓名
            student_id = re.search('^([a-zA-Z0-9]+)_', file).group(1)

            # 讀取xlsx檔案，讀取return欄位
            file_path = os.path.join(folder_path, file)
            df = pd.read_excel(file_path)
            values = df['return'].tolist()
            print(f"Student ID {student_id} with answers {values}")

            # 計算均方誤差
            mse = ((pd.Series(values) - pd.Series(correct_values)) ** 2).mean()

            # 儲存均方誤差
            accuracy_scores[student_id] = mse
        elif len(xlsx_files) > 1:
            print(f"Error: More than one .xlsx file found in folder {folder}. Skipping this folder.")
        else:
            print(f"Warning: No .xlsx file found in folder {folder}. Skipping this folder.")
```

```python
# 根據準確度排名
ranked_results = sorted(accuracy_scores.items(), key=lambda x: x[1])

# 印出排名結果
for rank, (student_id, score) in enumerate(ranked_results, start=1):
    print(f"Rank {rank}: Student ID {student_id} with accuracy {score:.6f}")
```
